```python
import pandas as pd
```

```python
df = pd.read_csv("LinkedIn_Post.csv", encoding='latin1')
df.shape
df.head(2)
```

```python
df.describe()
```

```python
df.info()
```

```python
keep_cols = [
    "postedDate", "contentType", "text", "totalReactionCount", 
    "commentsCount", "resharedPost"
]
df = df[keep_cols]

df.head(5)
```

```python
df.describe()
```

```python
df.info()
```

```python
df["postedDate"] = pd.to_datetime(
    df["postedDate"].str.replace(" \+0000 UTC", "", regex=True),
    format="%Y-%m-%d %H:%M:%S.%f"
)
```

```python
df["day_of_week"] = df["postedDate"].dt.day_name()
df["hour"] = df["postedDate"].dt.hour
```

```python
cols_to_numeric = ["totalReactionCount", "commentsCount", "resharedPost"]
for col in cols_to_numeric:
    df[col] = pd.to_numeric(df[col], errors="coerce").fillna(0)


df["engagement"] = (
    df["totalReactionCount"] +
    df["commentsCount"] +
    df["resharedPost"]
)
```

```python
df.head(5)
```

```python
df.to_csv("linkedin_cleaned.csv", index=False)

print("✅ Analysis complete. Charts displayed & cleaned file exported.")
```
